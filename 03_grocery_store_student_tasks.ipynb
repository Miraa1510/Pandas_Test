## Task 44

Use apply() to categorize Unit_Price into Low, Medium, High.

```python
df['Unit_Price'].min()
```

```python
df['Unit_Price'].max()
```

```python
def price_band(x):
    if x < df['Unit_Price'].quantile(0.22):
        return 'Low'
    elif x < df['Unit_Price'].quantile(0.55):
        return 'Medium'
    return 'High'

df['Price_Band'] = df['Unit_Price'].apply(price_band)
df['Price_Band'].head()
```
